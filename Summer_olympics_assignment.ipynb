### 1. In how many cities Summer Olympics is held so far?

```python
print('The number of different cities in which summer olympics is held so far is ',len(df['City'].unique()))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
gold_medals=[]
for i in df['Sport'].unique():
  gold_medals.append([i,len(df[(df['Medal'] == 'Gold') & (df['Sport'] == i)])])
gold_medals
```

```python
pd.DataFrame(gold_medals,columns = ['Sport','gold_medals']).sort_values(by='gold_medals', ascending=False).head().plot(x='Sport',y='gold_medals',kind='bar',color='#bd2c22',figsize=(10,5))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
medals=[]
for i in df['Sport'].unique():
  medals.append([i,len((df['Medal'])[df['Sport']  == i])])
medals
```

```python
pd.DataFrame(medals,columns = ['Sport','medals']).sort_values(by='medals', ascending=False).head().plot(x='Sport',y='medals',kind='bar',color='#21b82b',figsize=(10,5))
```

### 4. Which player has won most number of medals? (Top 5)

```python
medal=[]
for i in df['Athlete'].unique():
  medal.append([i,len((df['Medal'])[df['Athlete']  == i])])
medal
```

```python
pd.DataFrame(medal,columns = ['Athlete','medal']).sort_values(by='medal', ascending=False).head().plot(x='Athlete',y='medal',kind='bar',color='#21b82b',figsize=(10,5))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
medal_gold=[]
for i in df['Athlete'].unique():
  medal_gold.append([i,len(df[(df["Medal"] == "Gold") & (df["Athlete"] == i)])])
medal_gold
```

```python
pd.DataFrame(medal_gold,columns = ['Athlete','medal_gold']).sort_values(by='medal_gold', ascending=False).head().plot(x='Athlete',y='medal_gold',kind='bar',color='#9563e0',figsize=(10,5))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
a=df[(df['Country']=='IND') & (df['Medal']=='Gold')].reset_index()
print('The year in which India won first Gold Medal in Summer Olympics was ',a['Year'][0])
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
popular = []

for i in df['Event'].unique():
    popular.append([i , len(df[df['Event']  == i])])
popular
```

```python
pd.DataFrame(popular,columns = ['Event','popular']).sort_values(by='popular', ascending=False).head().reset_index().plot(x='Event',y='popular',kind='bar',color='#f2bf30',figsize=(10,5))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
female_gold_medal = []

for i in df['Sport'].unique():
    female_gold_medal.append([i , len(df[(df['Medal'] == 'Gold') & (df['Gender']=='Women') & (df['Sport']==i) ] ) ])
female_gold_medal
```

```python
pd.DataFrame(female_gold_medal,columns = ['Sport','female_gold_medal']).sort_values(by='female_gold_medal', ascending=False).head().plot(x='Sport',y='female_gold_medal',kind='bar',color='#a31a83',figsize=(10,5))
```
